```python
ulsan_answer = pd.read_csv('./data/ulsan_answer', sep=',').iloc[:,1:]

ulsan_answer.columns = ulsan_answer.columns.str.replace('ulsan_answer.', '')

ulsan_answer = ulsan_answer.applymap(lambda x: re.sub(r'"', '', str(x)))
```

```python
ulsan_answer
```

```python
# 정규표현식 패턴 정의
#위경도에서 N, E 제거
pattern = r'(\d+\.\d+)'

def extract_number(string):
    match = re.findall(pattern, string)
    if match:
        return float(match[0])
    else:
        return None
```

```python
ulsan_answer['lon'] = ulsan_answer['lon'].apply(extract_number)
ulsan_answer['lat'] = ulsan_answer['lat'].apply(extract_number)

ulsan_answer = ulsan_answer.drop_duplicates(subset=ulsan_answer.columns, keep='first').reset_index(drop=True)
```

```python
#busan_answer의 time컬럼 생성
#time column을 년,월,일,시간.분으로 나누는 함수
def split_time_colums(df, time_column):
    df[time_column] = pd.to_datetime(df[time_column])
    
    df['year'] = df[time_column].dt.year.apply(int)
    df['mon'] = df[time_column].dt.month.apply(int)
    df['day'] = df[time_column].dt.day.apply(int)
    df['hour'] = df[time_column].dt.hour.apply(int)
    df['min'] = df[time_column].dt.minute.apply(int)
    
    return df
```

```python
#busan_drag_train_final column형식을 busan_answer와 똑같이 맞춰주기
df1 = split_time_colums(ulsan_train,'time')
df1['num'] = df1['num'].apply(int)

#위경도 lon, lat으로 맞추기
df1.rename(columns = {'latitude':'lat', 'longitude' : 'lon'},inplace=True)
df1
```

```python
#ulsan_answer time객체 생성
def create_time(row):
    year=row['year']
    month = row['mon']
    day = row['day']
    hour = row['hour']
    minute = row['min']
    return pd.Timestamp(year=year, month=month, day=day, hour=hour, minute=minute)

df2['time'] = df2.apply(create_time, axis=1)
```

## df1과 df2매칭시켜서 time의 앞, 뒤 30분 target 1로

```python
df1['target']=0

def set_target_within_time_diff(row):
    time = row['time']
    num = row['num']
    
    target_rows = df1[(df1['num']==num)&(df1['time'] >= time - pd.Timedelta(minutes=30))&(df1['time'] <= time + pd.Timedelta(minutes=30))]
    df1.loc[target_rows.index, 'target']=1
    
df2.apply(set_target_within_time_diff, axis=1)
df1
```
